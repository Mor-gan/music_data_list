```python

#### Install all necessary libraries
import openpyxl
import re
from collections import Counter

#### Tidy data using regular expression 
def clean_string(s):
    return re.sub(r'[^a-zA-Z\s]', '', str(s)) if s else ''

#### Replace 'path/to/your/file.xlsx' with the actual full path to your Excel file
excel_file_path = '/Users/bimbo/Desktop/AbimbolaOgundipe_python/Music_list.xlsx'
workbook = openpyxl.load_workbook(excel_file_path)

#### Define the sheet name as a variable 'Sheet1', adjust it based on your actual sheet name
sheet = workbook['Sheet1']

#### Select the column containing song names, mine is column A, adjust it accordingly
song_names_column = 'A'

#### Extract only the names of the songs excluding invalid data (not numbers)
song_names = [clean_string(cell.value) for cell in sheet[f'{song_names_column}']]

# Remove empty strings and lowercase the song names for consistency
song_names = [name.lower() for name in song_names if name.strip()]

# Use Counter to count occurrences of each cleaned song name
song_name_counts = Counter(song_names)

# Get the most common names and their counts
most_common_names = song_name_counts.most_common()

# Display the result
print("Most Common Song Names:")
for name, count in most_common_names:
    print(f"{name}: {count} times")

#### My result shows 
#### Most Common Song Names:
#### amazing grace: 3 times
#### forever: 3 times
#### let it go: 3 times
#### intro: 3 times
#### lights out: 3 times
#### trouble: 3 times
```
